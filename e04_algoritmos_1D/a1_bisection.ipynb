# Descripción

- El máximo o mínimo de una función es donde $f'(x) = 0$.
- Debido a que la función es convexa (unimodal), la derivada tiende a cero conforme se acerca al mínimo global.
- Que una función sea unimodal significa que solo tiene un mínimo.
- El método de la bisección encuentra donde la derivada se hace $0$, es decir, la raíz de la función de la derivada.
- El método de la derivada elimina regiones. Dados los límites de la función $a$ y $b$, se calcula $\alpha$, el cual es la mitad de $a$ y $b$, posteriormente se evalúa la derivada en el punto $\alpha$, y se define cuál región debe ser eliminada.
-La región eliminada debe ser la que no contiene el mínimo, y eso se sabe porque hay un cambio de signo de la derivada. Matemáticamente se define como $f'(b)f'(a) < 0$ y los nuevos límites serán $a$ y $b = \alpha$, o $a = \alpha$ y $b$.

# Implementación

```python
from matplotlib import pyplot as plt
from time import time
import numpy as np
```

```python
# Función de costo del problema.
f_u = lambda t: ((204_165.5) / (330 - 2 * t)) + (10_400 / (t - 20))

# Definición de la derivada adelantada en un punto.
dx = lambda x, delta: (f_u(x + delta) - f_u(x)) / delta

# Restricciones de la variable t
a = 40
b = 90

# Tolerancia
epsilon = 0.001
delta_x = epsilon

# Registra tiempo de inicio del algoritmo.
start_time = time()

# Realiza las iteraciones del algoritmo
for x in range(1, 5000):
    alpha = (a + b) / 2
    fp_a = dx(a, delta_x)
    fp_alpha = dx(alpha, delta_x)
    
    if fp_a * fp_alpha < 0:
        b = alpha
    else:
        a = alpha
        
    if abs(dx(a, delta_x) - dx(b, delta_x)) <= epsilon:
        break
        
# Registra tiempo de finalización del algoritmo.
end_time = time()

# Calcula tiempo transcurrido en milisegundos.
elapsed_time = time() - start_time
elapsed_time = elapsed_time * 1000

print("Tiempo transcurrido: %f ms." % elapsed_time )
print("x: %f" % alpha)
print("fx: %f" % f_u(alpha))
print("f'x: %f" % fp_alpha)
print("Iteraciones: %d" % x)
```
